```python
from spacy.tokens import Span
from spacy.language import Language

# Register your custom component
@Language.component("edit_ents")
def edit_ents(doc):
    new_ents = []
    for ent in doc.ents:
        f = 0
        if any(x in ent.text.lower() for x in ["month", "day", "year", "annual"]):
            new_ent = Span(doc, ent.start, ent.end, label="CARDINAL")
            new_ents.append(new_ent)
            f = 1
        if f == 0:
            if ent.text.lower().split()[0] in nlp.Defaults.stop_words:
                new_ent = Span(doc, ent.start + 1, ent.end, label=ent.label)
                new_ents.append(new_ent)
            elif ent.text == "boycotts":
                pass
            else:
                new_ents.append(ent)
    doc.ents = new_ents
    return doc

# Add the registered component to the pipeline
nlp.add_pipe("edit_ents", after="ner")

# Process your text
doc = nlp(text)
[(ent.text, ent.label_) for ent in doc.ents]
```

```python
#algorithm for conversion to table
table={} #declaration of empty dictionary
rowno=0 #variable for setting each row
check_row_fill=0 #flag for ckecking if each row is filled up
#travering through all entities
for t in doc.ents:
    if check_row_fill==0: #if no row is filled up, we need to create a new row
        table[rowno]=[None]*3
    if t.label_ == 'DATE': #if label is a date
        if table[rowno][0] is not None:
            rowno=rowno+1
            check_row_fill=0
            table[rowno]=[None]*3
        table[rowno][0]=t.text
        check_row_fill=check_row_fill+1
    if t.label_ == 'PERSON': #if label is a person
        if table[rowno][1] is not None:
            rowno=rowno+1
            check_row_fill=0
            table[rowno]=[None]*3
        table[rowno][1]=t.text
        check_row_fill=check_row_fill+1
    if t.label_ == 'LAW' or t.label_ == 'EVENT': #if label is a event or law
        if table[rowno][2] is not None:
            rowno=rowno+1
            check_row_fill=0
            table[rowno]=[None]*3
        table[rowno][2]=t.text
        check_row_fill=check_row_fill+1
    if check_row_fill == 3:
        rowno=rowno+1
        check_row_fill=0
```

```python
table #printing the table dictionary
to_be_deleted =[]
for key in table.keys():
    if table[key][0] is None or table[key][2] is None:
        to_be_deleted.append(key)
        continue
    if table[key][1] is None and table[key][2] is None:
        to_be_deleted.append(key)
        continue
for d in to_be_deleted:
    del table[d]
table
```

```python
#converting dictionary into a dataframe
final_table=pd.DataFrame.from_dict(table).T
#renaming the column names
final_table.columns = ["DATE", "PERSON", "EVENT"]
final_table.reset_index(inplace=True)
final_table.drop('index',axis=1,inplace=True)
```
